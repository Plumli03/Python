```python
import pandas as pd

# Define the sample DataFrame
sample_df = pd.DataFrame({
    "sub": [50, 51, 52, 53, 54, 55, 56],
    "parent": [54, 55, 54, 56, 51, 0, 0]
})
# print(sample_df.info())

def find_ancestor_chain(df, sub):
    '''[topmost,level]   #print(find_ancestor_chain.__doc__) '''
    chain = []
    current = sub
    while True:
        chain.append(current)
        parent = df[df['sub'] == current]['parent'].values[0]
        if parent == 0 or parent == current:
            break
        current = parent
    # Find the top-most ancestor
    top_most_ancestor = max(chain)
    return [top_most_ancestor, len(chain) - 1]

# Example 
print(find_ancestor_chain(sample_df, 55))  # Output: [55, 0]
print(find_ancestor_chain(sample_df, 53))  # Output: [55, 3]
```
